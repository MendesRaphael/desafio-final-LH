```python
# Convertendo o nome das colunas para o nome que a biblioteca espera
Y_df = Y_df.rename({'total_qtd_product': 'y', 'orderdate': 'ds','productmodel_name':'product','name_store':'store','countryregioncode':'region'}, axis=1)
# Adicionando o nível de hierarquia maior, o país
Y_df.insert(0, 'Total', 'Total')
```

```python
Y_df.reset_index(drop=True, inplace=True)
```

```python
Y_df.head()
```

```python
print(Y_df.columns)
```

```python
# create unique identifiers per level in the hierarchy
Y_df['region']  = Y_df['Total'] + '>' + Y_df['region']
Y_df['store']   = Y_df['region'] + '>' + Y_df['store']
Y_df['product'] = Y_df['store'] + '>' + Y_df['product']

Y_df.head(3)
```

```python
# create the long format matrix: individual products
df_prod = Y_df.groupby(['ds', 'product'])[['y']].sum()
df_prod.reset_index(inplace=True)
df_prod = df_prod.T.reset_index(drop=True).T
df_prod.columns = ['ds', 'unique_id', 'sales']

# create the long format matrix: store level
df_store = Y_df.groupby(['ds', 'store'])[['y']].sum()
df_store.reset_index(inplace=True)
df_store.columns = ['ds', 'unique_id', 'sales']

# create the long format matrix: region level
df_reg = Y_df.groupby(['ds', 'region'])[['y']].sum()
df_reg.reset_index(inplace=True)
df_reg.columns = ['ds', 'unique_id', 'sales']

# create the long format matrix: total level
df_tot = Y_df.groupby(['ds'])[['y']].sum()
df_tot.reset_index(inplace=True)
df_tot['unique_id'] = 'Total'
df_tot.columns = ['ds', 'sales','unique_id']

# combine all four
dfx = pd.concat([df_prod, df_store, df_reg, df_tot], axis = 0)
print(df_prod.shape, df_store.shape, df_reg.shape, df_tot.shape, dfx.shape)

# format
xset = set(dfx.unique_id)
dfx.columns = ['ds','unique_id', 'y']
dfx['ds'] = pd.to_datetime(dfx['ds'])
```

```python
xset = set(dfx.unique_id)
list1 = list(xset)
list1.sort()

list2 = [f for f in list1 if f.count('>') == 3]
list2.sort()

S = np.zeros((len(list1), len(list2)))
S = pd.DataFrame(S); S.index = list1; S.columns = list2

# populate the summing matrix
S.loc['Total'] = 1
for xcol in list1[:-1]:
    fl = [f for f in list2 if xcol in f]
    S.loc[xcol][fl] = 1
```

```python
# encode the tags (structure per hierarchy level)
tags = {}
tags['Total'] = df_tot.unique_id.unique()
tags['Total/Region'] = df_reg.unique_id.unique()
tags['Total/Region/Store'] = df_store.unique_id.unique()
tags['Total/Region/Store/Product'] = df_prod.unique_id.unique()
```

```python
S_df = S[S.index.isin(x_hat.index)]
```

```python
# Reconcile the base predictions
reconcilers = [
    BottomUp(),
    TopDown(method='forecast_proportions'),
    MiddleOut(middle_level='Total/Region/Store',
              top_down_method='forecast_proportions'),
    MiddleOut(middle_level='Total/Region',
              top_down_method='forecast_proportions')
]

hrec = HierarchicalReconciliation(reconcilers=reconcilers)
# x_hat_rec = hrec.reconcile(x_hat, x_train, S, tags)
x_hat_rec = hrec.reconcile(x_hat, S_df, tags)
```
